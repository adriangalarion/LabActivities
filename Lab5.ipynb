```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
C = np.array([1,3])
print(C.ndim)
b = np.arange(0,30,3)
plt.xlim(6,30)
plt.ylim(10,30)
plt.plot(b*C[0], b*C[0], color='black') 
plt.plot(b*C[1], b*C[1], color='k') 
plt.show()
```

```python
C = np.array([
    [8,1,3],
    [2,1,3],
    [1,1,7]
])
fig = plt.figure()
A1 = plt.subplot(projection='3d')
A1.set_xlim([-4, 4])
A1.set_ylim([-4, 4])
A1.set_zlim([-4, 4])
origin = (-4,-4,-4)
A1.quiver(origin, origin, origin, C[:,0], C[:,1], C[:,2], arrow_length_ratio=0.07, colors=['k','red','yellow'])
plt.show()
```
